# **Previsão de Eleição para Deputados**

Nesta análise faremos uso de dados de campanhas passadas para prever se deputados serão eleitos em pleitos subsequentes.
As seguintes técnicas de aprendizagem de máquina serão aplicadas:

* KNN
* Árvores de Decisão
* Regressão Logística
* Adaboost

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from code import extract_x_y, plot_pie_imba, evaluate_model
from sklearn.model_selection import cross_val_score, train_test_split
#from sklearn.model_selection import train_test_split

%matplotlib inline
pd.set_option('display.max_columns', 200)
```

Os dados de treino utilizados são referentes aos anos de 2006 e 2010 para validação, enquanto que para o teste, utilizaremos dados da eleição de 2014.

```python
# Loading and preprocessing the data
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')

# Removing useless columns
train_data = train_data.drop(["sequencial_candidato", "nome"], axis = 1)
test_data = test_data.drop(["sequencial_candidato", "nome"], axis = 1)

# One Hot Encoding
train_features, train_target = extract_x_y(train_data)
train_features = pd.get_dummies(train_features)
train_data = pd.concat([train_features, train_target], axis=1)

train_features, validation_features, train_target, validation_target = train_test_split(train_features, train_target, test_size = .2)

# Separating training data for validation
#validation_data = train_data[train_data.ano == 2010]
#train_data = train_data[train_data.ano == 2006]

#train_features, train_target = extract_x_y(train_data)
```

OBS: Aqui fazemos uso da técnica de _One Hot Encoding_ para mapear variáveis categóricas em numéricas, possibilitanto desta forma o treinamento

## **1 - Desbalanceamento das Classes**

Em aprendizagem de máquina, quando possuímos dados muito desbalanceados (isto é, as classes não estão igualmente representadas nos dados de treino), alguns modelos podem aprender a tendêndia de dar preferência às categorias mais representadas. Como partimos do pressuposto de que possuímos apenas uma amostra dos dados, não sabemos se esta distribuição reflete a distribuição na população.

Vamos então tentar detectar esta propriedade nos nossos dados.

```python
plot_pie_imba(train_data)
```

Como já era de se esperar, com algum conhecimento do processo eleitoral, há muito mais candidatos não eleitos do que o contrário.

O problema de desbalanceamento pode ser amenizado utilizando a técnica de *Oversampling* conhecida como _SMOTE_. Nela, instâncias sintéticas da classe minoritária são geradas para equalizar um pouco as distribuições.

Entretanto, devemos levar em consideração os riscos supracitados, porém a desigualdade da representação em si, pode ser encarada como uma _feature_, então devemos buscar diminuir a diferença de representação mas não eliminá-la.

```python
#Oversampling using SMOTE
from imblearn.over_sampling import SMOTE

sm = SMOTE(ratio = .75, random_state = 42)

x_res, y_res = sm.fit_sample(train_features, train_target)
balanced_train_data = pd.concat([pd.DataFrame(x_res), pd.DataFrame(y_res)], axis=1)
balanced_train_data.columns = train_data.columns
```

```python
plot_pie_imba(balanced_train_data)
```

Como podemos perceber, as diferenças agora estão atenuadas.

## **2 - Treinamento de Modelos**

Vamos então iniciar o treinamento dos modelos pela família de modelos KNN.

```python
train_features, train_target = extract_x_y(balanced_train_data)
```

### **2.1 - KNN**

Faremos uso de validação cruzada para selecionar o melhor valor para o hiperparâmetro K.

```python
from sklearn.neighbors import KNeighborsClassifier

k_range = list(range(1, 21))

knn_scores = list(map(lambda x: cross_val_score(KNeighborsClassifier(n_neighbors=x), train_features, train_target, cv=10, scoring="f1_macro").mean(),
    k_range))

plt.plot(k_range, knn_scores)
plt.xlabel('Valor de K')
plt.xticks(k_range)
plt.ylabel('F1')
```

De acordo com a visualização escolheremos o valor 3 para manter o modelo simples e performático.

```python
final_knn_classifier = KNeighborsClassifier(n_neighbors = 3).fit(train_features, train_target)
```

## **2.2 - Regressão Logística**

```python
from sklearn.linear_model import LogisticRegression

c_range = [0.1, 1, 100]
reg_log_scores = list(map(lambda x: cross_val_score(LogisticRegression(solver = 'lbfgs', C = x, max_iter=1000), train_features, train_target, cv=10, scoring="f1_macro").mean(),
    c_range))
```

```python
plt.plot(c_range, reg_log_scores)
plt.xlabel('Valor de C')
plt.xticks(c_range)
plt.ylabel('F1')
```

Com os valores para a métrica, escolheremos 100.

```python
final_log_reg_classifier = LogisticRegression(solver = 'lbfgs', C = 100, max_iter=1000).fit(train_features, train_target)
```

## **2.3 - Árvores de Decisão**

```python
from sklearn.tree import DecisionTreeClassifier

depth_range = list(range(1, 25, 5))

tree_scores = list(map(lambda x: cross_val_score(DecisionTreeClassifier(max_depth=x), train_features, train_target, cv=10, scoring="f1_macro").mean(),
    depth_range))

```

```python
plt.plot(depth_range, tree_scores)
plt.xlabel('Profundidade da Árvore')
plt.xticks(depth_range)
plt.ylabel('F1')
```

A profundidade ótima para o nosso problema, é de 11.

```python
final_tree_classifier = DecisionTreeClassifier(max_depth=11).fit(train_features, train_target)
```

## **2.4 - Adaboost**

```python
from sklearn.ensemble import AdaBoostClassifier

estimators_range = [50, 100, 150, 250, 400, 610, 900]

adaboost_scores = list(map(lambda x: cross_val_score(AdaBoostClassifier(n_estimators=x), train_features, train_target, cv=10, scoring="f1_macro").mean(),
    estimators_range))
```

```python
plt.plot(estimators_range, adaboost_scores)
plt.xlabel('Número de Estimadores')
plt.xticks(estimators_range)
plt.ylabel('F1')
```

Selecionamos então 250 como número de estimadores.

```python
final_adaboost_classifier = AdaBoostClassifier(n_estimators=250).fit(train_features, train_target)
```

Agora que já treinamos nossos modelos. Vamos avaliá-los pelas métricas:
* Precisão

* Área sob a Curva ROC (AUC)
